```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the clinicaltrials.gov search page
url = 'https://clinicaltrials.gov/ct2/results?'
#cond=&term=&cntry=&state=&city=&dist=
# Set the phase parameter to 2 to search for phase 3 clinical trials (setting this to 3 brings back phase 4)
# Set the cntry parameter to 'US' to search for clinical trials in the United States
url += '&phase=2&cntry=US'

# Make the HTTP request to the URL
response = requests.get(url)

# Parse the HTML content of the page
soup = BeautifulSoup(response.text, 'html.parser')

# Create an empty list to store the data for each clinical trial
trials_data = []

# Find all the rows in the search results table
rows = soup.find_all('tr')

# Iterate through each row
for row in rows:
  # Find the NCTId cell in the row
  nct_cell = row.find('td', {'class': 'nct'})
  if nct_cell:
    # Extract the NCTId from the cell
    nct_id = nct_cell.text.strip()
    
    # Find the title cell in the row
    title_cell = row.find('td', {'class': 'title'})
    # Extract the title from the cell
    title = title_cell.text.strip()
    
    # Find the status cell in the row
    status_cell = row.find('td', {'class': 'status'})
    # Extract the status from the cell
    status = status_cell.text.strip()
    
    # Add the data for this clinical trial to the list
    trials_data.append({
      'NCTId': nct_id,
      'Title': title,
      'Status': status
    })

# Create a Pandas dataframe from the trials data
df = pd.DataFrame(trials_data)

# Print the dataframe
print(df)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the clinicaltrials.gov search page
url = 'https://clinicaltrials.gov/ct2/results?cond=&term=&cntry=&state=&city=&dist='

# Set the phase parameter to 2 to search for phase 2 clinical trials
# Set the cntry parameter to 'US' to search for clinical trials in the United States
url += '&phase=2&cntry=US'

# Make the HTTP request to the URL
response = requests.get(url)

# Parse the HTML content of the page
soup = BeautifulSoup(response.text, 'html.parser')

# Find the data table on the page
table = soup.find('table', {'id': 'theDataTable'})
print(table)

# Find all the rows in the table
rows = table.find_all('tr')

# Create an empty list to store the data for each clinical trial
trials_data = []

# Iterate through each row
for row in rows:
  # Find the cells in the row
  cells = row.find_all('td')
  if cells:
    # Extract the data from the cells
    nct_id = cells[0].text.strip()
    title = cells[1].text.strip()
    status = cells[2].text.strip()
    
    # Add the data for this clinical trial to the list
    trials_data.append({
      'NCTId': nct_id,
      'Title': title,
      'Status': status
    })

# Create a Pandas dataframe from the trials data
df = pd.DataFrame(trials_data)

# Print the dataframe
print(df)
```
